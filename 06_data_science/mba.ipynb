```python
import matplotlib.pyplot as plt

labels = ['18–25 Years', '26–35 Years', '36–45 Years', 'Above 45 Years']
sizes = [40, 35, 15, 10]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(8,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=(0.05, 0, 0, 0))
plt.title('Age-wise Distribution of Respondents', fontsize=14, fontweight='bold')
plt.axis('equal') 
plt.savefig('age_distribution.png', dpi=300)
plt.show()
```
